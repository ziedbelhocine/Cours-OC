```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

prets = pd.read_csv('prets.csv')
prets['taux_endettement'] = round(prets['remboursement'] * 100 / prets['revenu'], 2)
prets.head()
```

```python
import seaborn as sns
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type')
```

```python
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', size='duree')
```

```python
plt.figure(figsize=(10,6))


plt.rcParams.update({'font.size': 14})


sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type')


plt.ylabel("Taux d'endettement")


plt.xlabel('Revenu mensuel ()')


plt.grid()


plt.xlim(500, 7500)


plt.legend(bbox_to_anchor=(1, 1.02))


plt.title("Taux d'endettement en fonction du revenu mensuel\npar type de prt contract")


plt.show()
```

```python
sns.barplot(data=prets, x='ville', y='remboursement', errorbar='ci', estimator=sum)
```

```python
sns.set_palette('pastel')
sns.barplot(data=prets, x='ville', y='remboursement', errorbar='ci', estimator=np.mean, hue='type')
plt.legend(loc='upper right', fontsize=8)
```

```python
sns.set_palette('dark')
sns.lineplot(data=prets, x='ville', y='remboursement', errorbar='ci', estimator=np.mean, hue='type')
plt.legend(loc='upper right', fontsize=8)
```

```python
sns.set_palette('Set2')
sns.boxplot(data=prets, x='ville', y='remboursement', hue='type')
plt.legend(loc='upper right', fontsize=8)
```

```python
sns.set_palette('Set2')
sns.histplot(data=prets, x='ville', y='remboursement')
```
